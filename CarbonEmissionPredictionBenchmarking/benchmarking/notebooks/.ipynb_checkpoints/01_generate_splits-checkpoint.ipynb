```python
import sys
import os
import numpy as np
sys.path.append(os.path.abspath("../.."))

from benchmarking.config.global_config import *
from benchmarking.utils.preprocessing_utils import (
    load_and_prepare_data,
    split_by_ratio,
    get_feature_columns,
    scale_numeric_features,
    create_sequences
)
```

```python
import random
import tensorflow as tf

random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)
```

```python
DATA_PATH = r"D:\MUJ\Projects\CarbonEmissionPredictionRNN-NinOA\CarbonEmissionPredictionBenchmarking\data\raw\2agosto -dic 2021.csv"
```

```python
for k in HORIZONS:

    print(f"\n Processing horizon: {k}")

    # 1. Load and preprocess
    df = load_and_prepare_data(DATA_PATH, k)

    print("Data shape after preprocessing:", df.shape)

    # 2. Split
    train, val, test = split_by_ratio(df)

    print("Train:", train.shape, "Val:", val.shape, "Test:", test.shape)

    # 3. Feature columns
    feature_cols = get_feature_columns(df)

    # 4. Scale
    train, val, test, scaler = scale_numeric_features(
        train, val, test, feature_cols
    )

    # 5. Create sequences
    X_train, y_train = create_sequences(
        train[feature_cols].values,
        train[TARGET_COLUMN].values,
        SEQUENCE_LENGTH
    )

    X_test, y_test = create_sequences(
        test[feature_cols].values,
        test[TARGET_COLUMN].values,
        SEQUENCE_LENGTH
    )

    print("Sequence shapes:")
    print("X_train:", X_train.shape, "y_train:", y_train.shape)
    print("X_test:", X_test.shape, "y_test:", y_test.shape)

    # 6. Save splits
    save_path = os.path.join(SPLIT_PATH, f"horizon_{k}")
    os.makedirs(save_path, exist_ok=True)

    np.save(os.path.join(save_path, "X_train.npy"), X_train)
    np.save(os.path.join(save_path, "y_train.npy"), y_train)
    np.save(os.path.join(save_path, "X_test.npy"), X_test)
    np.save(os.path.join(save_path, "y_test.npy"), y_test)
    np.save(os.path.join(save_path, "X_val.npy"), X_val)
    np.save(os.path.join(save_path, "y_val.npy"), y_val)

    print("Saved successfully at:", save_path)
```
