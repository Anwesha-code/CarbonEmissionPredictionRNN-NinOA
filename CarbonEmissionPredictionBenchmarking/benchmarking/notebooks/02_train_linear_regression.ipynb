```python
import numpy as np
import os

from benchmarking.models.classical.linear_regression import LinearRegressionModel
from benchmarking.utils.metrics import evaluate_model
```

```python
HORIZONS = [60, 90, 300, 1800]
BASE_SPLIT_PATH = "results/splits"

all_results = []

for k in HORIZONS:
    print(f"\n--- Training for Horizon: {k} ---")

    split_path = os.path.join(BASE_SPLIT_PATH, f"horizon_{k}")

    X_train = np.load(os.path.join(split_path, "X_train.npy"))
    y_train = np.load(os.path.join(split_path, "y_train.npy"))
    X_test = np.load(os.path.join(split_path, "X_test.npy"))
    y_test = np.load(os.path.join(split_path, "y_test.npy"))
    
    print("Train shape:", X_train.shape)
    print("Test shape:", X_test.shape)

    model = LinearRegressionModel()
    model.fit(X_train, y_train)

    results_df, _, _, _ = evaluate_model(model, X_test, y_test)

    results_df["Model"] = "LinearRegression"
    results_df["Horizon"] = k

    all_results.append(results_df)
```

```python
results_df, y_true, y_pred, explanation = evaluate_model(
    model, X_test, y_test
)

print(results_df)
print("\nExplanation:", explanation)
```

```python
import pandas as pd

SAVE_PATH = "results/analysis"
os.makedirs(SAVE_PATH, exist_ok=True)

results_df.to_csv(os.path.join(SAVE_PATH, "linear_regression_results.csv"), index=False)
```
